```python
import json
import pandas as pd
import random
from openai import OpenAI

def get_random_rows(df, num_rows):
    """
    Selects a random number of rows (between 5-10) from the DataFrame.
    """
    num_to_sample = random.randint(5, num_rows)
    return df.sample(num_to_sample, random_state=random.randint(1, 1000))

def extract_key_value_pairs(row, client):
    """
    Sends the abstract text of a row to the OpenAI API to extract key-value pairs.
    """
    row_id = int(row["AwardID"])
    abstract = row["AbstractNarration"]
    
    messages = [
        {
            "role": "system",
            "content": (
                "You are an expert at extracting key value pairs from paragraphs. "
                "Please give valid output JSON"
            )
        },
        {
            "role": "user",
            "content": f"""
    Task: 
    Your task is to extract meaningful key-value pairs from a given piece of text. The keywords should be of both types - generalized (meaning these keys can be found in other similar texts) and unique (meaningful keywords uniquely found in the text). 
    Make sure that these key-value pairs cover all important aspects of the text in a short, concise manner. 

    Instructions for key-value pairs:
    1. Keys should be:
    - In snake_case format (words separated by underscores)
    - Descriptive but concise
    - Act like database column names or metadata tags
    - Should not be direct questions
    - Should clearly indicate what information they represent
    - Try to create a diverse set of keys
    - Try to include keys that are descriptive of the text
    - Be sure to have some keys that tell what exactly is the grant/prject about
    - Be sure to add the various fields covered by the project such as [computer science, biology, civil science, humanities] etc.


    2. Values should:
    - Directly correspond to information from the text
    - Be precise and complete
    - Include units or additional context when relevant
    - Values should be concise in nature
    - Values should correctly represent important information from the text
    - Simplify the language present in the values and extract the exact words in the sentence that are useful. They need not be grammatically correct just short. 

    # Text
    row_id = {row_id}
    {abstract}

    Please follow the Response format while answering :
        row_id: {{
            key: value
        }}
            """
            }
        ]

    chat_completion, *_ = client.chat.completions.create(
            model="gpt-4o-mini", 
            messages=messages,
            response_format={"type": "json_object"},
            temperature=0.4
    ).choices
    content = chat_completion.message.content
    return json.loads(content)

def process_multiple_rows(df, client, num_rows=10):
    """
    Processes multiple rows from the DataFrame and extracts key-value pairs.
    """
    rows = get_random_rows(df, num_rows)
    extracted_data = []

    for _, row in rows.iterrows():
        try:
            result = extract_key_value_pairs(row, client)
            extracted_data.append(result)
        except Exception as e:
            print(f"Error processing row {row['AwardID']}: {e}")

    return extracted_data

def update_dataframe_with_json(df, extracted_data):
    """
    Updates the DataFrame based on extracted key-value pairs.
    """
    for item in extracted_data:
        for row_id, key_values in item.items():
            row_id = int(row_id)
            if row_id not in df["AwardID"].values:
                print("row id not found")
                continue
            
            for key, value in key_values.items():
                if key not in df.columns:
                    df[key] = None  # Add new column with default None values
                
                # Update the specific row
                df.loc[df["AwardID"] == row_id, key] = value
    return df

def save_json_to_file(json_data, file_name="extracted_data.json"):
    """
    Saves extracted JSON data to a file.
    """
    with open(file_name, "w") as json_file:
        json.dump(json_data, json_file, indent=4)
    print(f"JSON data saved to {file_name}.")


def main():
    # Load the DataFrame
    df = pd.read_csv("sub_2008_data.csv", dtype=str)
    client = OpenAI()

    # Process multiple rows and extract JSON
    extracted_data = process_multiple_rows(df, client, num_rows=10)
    # Save the extracted JSON data to a file
    save_json_to_file(extracted_data, file_name="extracted_data.json")


    # Update the DataFrame with extracted JSON
    df = update_dataframe_with_json(df, extracted_data)

    # Save updated DataFrame (optional)
    df.to_csv("updated_data.csv", index=False)
    print("DataFrame updated and saved to 'updated_data.csv'.")

if __name__ == "__main__":
    main()
```
